```
import matplotlib.pyplot as plt
```

```
# This single perceptron will learn the linear function: y = 2x + 4
inp = [1, 2, 3, 4, 5, 6, 7, 8]
gt = [6, 8, 10, 12, 14, 16, 18, 20]
print([2*x + 4 for x in inp])
```

```
plt.plot(inp, gt, 'r-')
plt.show()
```

```
w1 = 1.5
w2 = 1.5
b = 1
eta = 0.1

epochs = 20
```

```
exp = []
for epoch in range(epochs):
    print("Epoch {}".format(epoch))
    for i, t in zip(inp, gt):
        neuron_output = i*w1 + b
        # if neuron_output >= 0:
        #     activation =1
        # else:
        #     activation = 0
        error = t - neuron_output

        w1 = w1 + eta*error*i
        b = b + eta*error

        x = (-(b/w1), 0)

        print("w1:", w1, "b:", b)
        print("x:", x)
        exp.append([w1*i + b for i in inp])
```

```
print(exp)
plt.plot(inp, gt, 'g-')
plt.plot(inp, exp[25], 'r-')
plt.show()
```

```
f, ((ax1, ax2, ax3), (ax4, ax5, ax6)) = plt.subplots(2, 3)
ax1.plot(inp, gt, 'g--')
ax1.plot(inp, exp[0], 'r-')
ax2.plot(inp, gt, 'g--')
ax2.plot(inp, exp[1], 'r-')
ax3.plot(inp, gt, 'g--')
ax3.plot(inp, exp[2], 'r-')
ax4.plot(inp, gt, 'g--')
ax4.plot(inp, exp[3], 'r-')
ax5.plot(inp, gt, 'g--')
ax5.plot(inp, exp[10], 'r-')
ax6.plot(inp, gt, 'g--')
ax6.plot(inp, exp[29], 'r-')
```

```
def predict(w1, b, inp):
    return inp*w1 + b
```
