```
import matplotlib.pyplot as plt
```

```
input = [(0, 0), (0, 1), (1, 0), (1, 1)]
gt = [0, 1, 1, 1]
w1 = 0.5
w2 = 0.5
b = 1
epochs = 20
```

```
x_plot = []
y_plot = []

for epoch in range(epochs):
    print("Epoch {}".format(epoch))
    for inp, t in zip(input, gt):
        neuron_output = inp[0]*w1 + inp[1]*w2 + b
        if neuron_output >= 0:
            activation =1
        else:
            activation = 0
        error = t - activation

        w1, w2 = w1 + error*inp[0], w2 + error*inp[1]
        b = b + error

        x1 = (-(b/w1), 0)
        x2 = (0, -(b/w2))

        print("w1", w1, "w2:", w2, "b", b)
        print("x1:", x1, "x2:", x2)
        
        x_plot.append([0, -b/w1])
        y_plot.append([-b/w2, 0])
```

```
# print(y_plot)
# print(x_plot)
x = [0, 0, 1, 1]
y = [0, 1, 0, 1]
plt.scatter(x[1:], y[1:], s=150)
plt.scatter(x[0], y[0], s=150)
plt.plot(x_plot[4], y_plot[4], 'r-')
plt.show()
```
